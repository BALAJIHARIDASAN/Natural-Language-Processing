```python
##tensorflow >2.0
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['Break long paragraphs up',
        'Go really short when you want to make a point',
        ' Shorten your sentence length',
        ' Review your word choices',
        'Use rhythm to guide your paragraphs',
        ' Look at your paragraphs on the page']
```

```python
### Vocabulary size
voc_size=5000
```

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

```python

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)  # padding - to make the length of sentence equal pre/post
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python

print(model.predict(embedded_docs))
```
