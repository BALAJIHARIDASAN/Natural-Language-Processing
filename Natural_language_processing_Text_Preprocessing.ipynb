```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aswintechguy/Data-Science-Concepts/main/NLP/data/Twitter%20Sentiments.csv')
```

```python
df.drop(['id'],axis = 1,inplace= True)
```

```python
df.head()
```

```python
# lower case

df['clean_text'] = df['tweet'].str.lower()
```

```python
df.head()
```

```python
# remove punctutations
import string
def remove_punc(text):
  punctuations= string.punctuation
  return text.translate(str.maketrans('','',punctuations))
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x:remove_punc(x))
```

```python
df.head()
```

```python
#stopword removal
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
STOPWORDS = set(stopwords.words('english'))
```

```python
def remove_stopwords(text):
  return " ".join([word for word in text.split()if word not in STOPWORDS])
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x:remove_stopwords(x))
```

```python
df.head()
```

```python
#removal of frequent words
```

```python
from collections import Counter
word_count = Counter()
for text in df['clean_text']:
  for word in text.split():
    word_count[word]+=1
```

```python
word_count.most_common(10)
```

```python
FREQUENT_WORDS = set(word for (word, wc) in word_count.most_common(3))
```

```python
def remove_freq_word(text):
  return " ".join([word for word in text.split()if word not in FREQUENT_WORDS])
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x:remove_freq_word(x))
```

```python
df.head()
```

```python
# remove rare words
RARE_WORDS = set(word for (word, wc) in word_count.most_common()[:-10:-1])
```

```python
def remove_rare_word(text):
  return " ".join([word for word in text.split()if word not in RARE_WORDS])
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x:remove_rare_word(x))
```

```python
df.head()
```

```python
# removal of special characters
import re
def removal_special_char(text):
  text = re.sub('[^a-zA-Z0-9]',' ',text)
  text = re.sub('\s+',' ',text)
  return text
```

```python
df['clean_text'] = df['clean_text'].apply(lambda x:removal_special_char(x))
```

```python
df.head()
```

```python
#stemming
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
df['stemmed_word'] = df['clean_text'].apply(lambda x:stem_words(x))
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()
wordnet_map = {"N":wordnet.NOUN, "V": wordnet.VERB, "J": wordnet.ADJ, "R": wordnet.ADV}

def lemmatize_words(text):
    # find pos tags
    pos_text = pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN))for word,pos in pos_text])
```

```python
nltk.download('averaged_perceptron_tagger')
df['lemmatized_text'] = df['clean_text'].apply(lambda x: lemmatize_words(x))
df.head()
```

```python
#removal of urls
def remove_url(text):
    return re.sub(r'https?://\S+|www\.\S+', '', text)
```

```python
#removal of html tags
def remove_html_tags(text):
    return re.sub(r'<.*?>', '', text)
```

```python
from spellchecker import SpellChecker
spell = SpellChecker()

def correct_spellings(text):
    corrected_text = []
    misspelled_text = spell.unknown(text.split())
    # print(misspelled_text)
    for word in text.split():
        if word in misspelled_text:
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)

    return " ".join(corrected_text)
```

```python
df['lemmatized_text'] = df['lemmatized_text'].apply(lambda x: remove_url(x))
df['lemmatized_text'] = df['lemmatized_text'].apply(lambda x: remove_html_tags(x))

```
